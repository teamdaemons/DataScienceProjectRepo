### 5.3 Processing

- Dealing with missing values <br/>
    - Replacing missing values of __Revenue (Millions)__ and __Metascore__ with median values.

```python
Revenue = imdb['Revenue (Millions)'].median()
Metascore = imdb.Metascore.median()
imdb['Revenue (Millions)'].fillna(Revenue, inplace = True)
imdb.Metascore.fillna(Metascore, inplace = True)
```

```python
imdb.isnull().sum()
```

- Creating a new column named __CategoryOnRating__, Categorising the 'Rating' column. if the Rating is less than or equal to 'min' that will be called as 'Not watchable', if rating is more than 'min' and less than or equal to '25%' that will be called as 'Not Good', if rating is more than '25%' and less than or equal to '50%' that will be called as 'Below Average', if rating is more than '50%' and less than or equal to '75%' that will be called as 'Above Average', if rating is more than '75%' and less than or equal to 'max' that will be called as 'Value for Money'.

- Note: Please refer the Describe output below to understand the parameters well.

```python
desc = imdb.describe()
desc
```

```python
def categ_on_rat(rate):
    if rate <= desc.loc['min']['Rating']:
        return 'Not watchable'
    elif rate > desc.loc['min']['Rating'] and rate <= desc.loc['25%']['Rating']:
        return 'Not Good'
    elif rate > desc.loc['25%']['Rating'] and rate <= desc.loc['50%']['Rating']:
        return 'Below Average'
    elif rate > desc.loc['50%']['Rating'] and rate <= desc.loc['75%']['Rating']:
        return 'Above Average'
    elif rate > desc.loc['75%']['Rating'] and rate <= desc.loc['max']['Rating']:
        return 'Value for Money'  
```

```python
imdb['CategoryOnRating'] = imdb['Rating'].apply(lambda x: categ_on_rat(x))
```

### 6.6 How strongly is ‘Rank’, ‘Votes’,  'Metascore' variables correlated with Rating ?

```python
imdb[['Rank', 'Votes', 'Metascore', 'Rating']].corr().style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

- We shall see by ploting scattered plot, how is the correlation between __'Rank', 'Votes', 'Metascore'__ and __'Rating'__    
  Variables.

```python
# plot correlated values
plt.rcParams['figure.figsize'] = [16, 6]

fig, ax = plt.subplots(nrows=1, ncols=3)

ax=ax.flatten()

cols = ['Rank', 'Votes', 'Metascore']
colors=['#415952', '#f35134', '#243AB5', '#243AB5']
j=0

for i in ax:
    if j==0:
        i.set_ylabel('Rating')
    i.scatter(imdb[cols[j]], imdb['Rating'],  alpha=0.5, color=colors[j])
    i.set_xlabel(cols[j])
    i.set_title('Pearson: %s'%imdb.corr().loc[cols[j]]['Rating'].round(2)+' Spearman: %s'%imdb.corr(method='spearman').loc[cols[j]]['Rating'].round(2))
    j+=1
```

#### Observation

- You can see in the above correlation table that __'Votes' and 'Metascore'__ are correlated to __'Rating'__ of the movies.
- You can see from the Pearson and Spearman scatter plot, that Linear change and Monotonic relation of the three 
  variables with __Rating__ of the movie.
- Appart from Ranking all other entities have correlation with __Rating__
- Leats use the heat map shown below to find the correlation between all the parameters in the data frame

<a id=section607><a/>

### 6.7 Establish coorelation between all the features using heatmap.

```python
corr = imdb.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,vmax=.8,linewidth=.01, square = True, annot = True,cmap='YlGnBu',linecolor ='black')
plt.title('Correlation between all the parameters')
```

#### Observation

- You can see that __Rating __, __Votes__, __Revenue (Millions)__ and __Metascore__ have positive correlation with __Runtime 
  (Minutes)__  and  __Revenue (Millions)__ and __Votes__, __Metascore__, __Rating __, __Runtime (Minutes)__  have good positive   Correlation.
- This shows that __Runtime (Minutes)__ of the Movie is an considerable entity which indirectly affects the __Votes__, 
  __Metascore__ and __Rating__ which In turn  affecting the __Revenue (Millions)__.
